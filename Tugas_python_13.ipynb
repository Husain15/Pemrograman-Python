```
import numpy as np
import matplotlib.pyplot as plt
```

```
def cos(frek,tipe,ht,hr,d):
  if frek in range(150,1501):
        Ca=69.55
        Cb=26.16
  elif frek in range(1501,2001):
        Ca=46.3
        Cb=33.9
    
  if tipe=='urban' :
        Cm=0
  elif tipe=='suburban' :
        Cm= -2*((np.log10((frek/28)**2))-5.4)
  elif tipe=='open' :
        Cm= -4.78*(np.log10(frek))+18.33*(np.log10(frek))-40.94
  else :
        Cm='tipe salah'

  ahr = (1.1*np.log10(frek)-0.7)*hr-(1.56*np.log10(frek)-1.8)
  Lp = Ca+(Cb*np.log10(frek))-(13.83*np.log(ht))-ahr+((44.9-6.55*np.log10(ht))*np.log10(d))-Cm
    
  return Lp
```

```
urban = []
open = []
suburban = []
```

```
for i in range(1, 202):
  i = i * 2.5
  x = 200 + i - 2.5
  pl1 = cos(1800, "urban", 80, 1.5, x)
  urban.append(pl1)
```

```
for i in range(1, 202):
  i = i * 2.5
  x = 200 + i - 2.5
  pl1 = cos(1800, "open", 80, 1.5, x)
  open.append(pl1)
```

```
for i in range(1, 202):
  i = i * 2.5
  x = 200 + i - 2.5
  pl1 = cos(1800, "suburban", 80, 1.5, x)
  suburban.append(pl1)
```

```
x = []
for i in range(1, 202):
  i = i * 2.5
  x.append(200 + i - 2.5) 
```

```
plt.figure()
plt.plot(x, urban, label="Urban")
plt.plot(x, open, label="Open")
plt.plot(x, suburban, label="Sub Urban")
plt.legend()
plt.title('Kurva Path loss beda daerah')
plt.xlabel('Jarak (M)')
plt.ylabel('Pathloss (dB)')
plt.show()
```
